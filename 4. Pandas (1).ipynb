### What is the number of columns in the dataset?

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
df = pd.read_csv(url, sep = '\t')

len(df)
````

````python
len(df.columns)
````

### Which was the most-ordered item and how many times it was ordered?

````python
print(df["item_name"].value_counts().idxmax())
print(df["item_name"].value_counts().max())
````

### How many different items are sold?

````python

df.groupby("item_name").sum()["quantity"].sort_values(ascending = False)
````

### How many items were orderd in total?

````python
len(df.groupby("item_name"))
````

### Turn the item price into a float

````python
df["item_clearprice"] = df["item_price"].apply(lambda x: float(x[1:]))
df.head(10)
````

### What is the price of each item?

````python
df[["item_name", "item_price"]].drop_duplicates().reset_index()
````

### How much was the revenue for the period in the dataset?

````python
(df["item_clearprice"] * df["quantity"]).sum()
````

### How many orders were made in the period?

````python
df["order_id"].value_counts().count()
````

### What is the average revenue amount per order?

````python
(df["item_clearprice"] * df["quantity"]).sum() / df["quantity"].sum()
````

### Merge the table ```dfa``` that contains order date and order id with ```df``` and plot daily revenue 

````python
dfa = pd.DataFrame({"date": np.random.choice(pd.date_range(start="20170101", end="20170201"), size=df.order_id.value_counts().index.size),
                   "order_id": np.arange(df.order_id.value_counts().index.size)})
dat = dfa["date"][df["order_id"] - 1]
dat.index = np.arange(dat.size)
df["date"] = dat
ans = df.groupby("date")["item_clearprice"].sum()
plt.plot(ans.index, ans.values)
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
````

### Create a histogram of the top 5 items bought

````python
plt.hist(df.item_name.value_counts()[:5], bins = 10)
plt.show()
````

### Sort the dataset by the name of the item

````python
df.sort_values(by = ['item_name'])
````

### How many times were a Veggie Salad Bowl ordered?

````python
df.loc[df["item_name"] == "Veggie Salad Bowl"].sum()["quantity"]
````

### How many times people orderd more than one Canned Soda? 

````python
df.loc[(df["item_name"] == "Canned Soda") & (df["quantity"] > 1)].size
````

### Create a scatterplot with the number of items orderered per order price

````python
p = df.groupby("order_id")["item_clearprice"].sum()
q = df.groupby("order_id")["quantity"].sum()
plt.scatter(p.values, q.values)
plt.xlabel("Price")
plt.ylabel("Number of items")
plt.show()
````
