```python
import numpy as np
arr = np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt")

import matplotlib.pyplot as plt

plt.plot(arr[:, 0], arr[:, 1], label = "Hares")
plt.plot(arr[:, 0], arr[:, 2], label = "Lynxes")
plt.plot(arr[:, 0], arr[:, 3], label = "Carrots")
plt.xticks(np.arange(arr[:, 0].min(), arr[:, 0].max(), 3))
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
ind = np.argpartition(arr[:, 1], 2)[:2]
print(ind)
for i in ind:
    plt.axvline(arr[i, 0])
ind = np.argpartition(arr[:, 2], 2)[:2]
print(ind)
for i in ind:
    plt.axvline(arr[i, 0], c = 'orange')
    ind = np.argpartition(arr[:, 3], 2)[:2]
print(ind)
for i in ind:
    plt.axvline(arr[i, 0], c = 'g')
plt.show()
```
